```python
import pandas as pd
import json
```

```python
file_path = '../data/miband/20260406_8278074507_MiFitness_ams1_data_copy/20260406_8278074507_MiFitness_hlth_center_aggregated_fitness_data.csv'
df_raw = pd.read_csv(file_path)
```

```python
df_steps = df_raw[(df_raw['Key'] == 'steps') & (df_raw['Tag'] == 'daily_report')].copy()

steps_data_list = [json.loads(val) for val in df_steps['Value']]
df_steps_normalized = pd.json_normalize(steps_data_list)

df_steps_result = df_steps[['Uid', 'Sid', 'Time']].reset_index(drop=True).join(df_steps_normalized)

df_steps_result
```

```python
df_steps_result['Time'] = pd.to_datetime(df_steps_result['Time'], unit='s')

df_steps_result
```

```python
import os

output_dir = '../data/processed_data'
output_path = os.path.join(output_dir, 'miband_steps_processed.csv')
df_steps_result.to_csv(output_path, index=False)
```
