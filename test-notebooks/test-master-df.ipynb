```python
import pandas as pd
import glob
import os
```

```python
# Get all CSV files from the processed_data directory
data_dir = '../data/processed_data'
csv_files = glob.glob(os.path.join(data_dir, '*.csv'))

# Read all CSV files
dfs_to_merge = []
for file in csv_files:
    df = pd.read_csv(file)
    # Ensure Date column is datetime
    if 'Date' in df.columns:
        df['Date'] = pd.to_datetime(df['Date'])
        dfs_to_merge.append(df)
    else:
        print(f"Warning: {os.path.basename(file)} has no 'Date' column, skipping")

print(f"\nLoaded {len(dfs_to_merge)} dataframes with 'Date' column")
```

```python
# Merge all dataframes using outer join on Date
if dfs_to_merge:
    merged_master_df = dfs_to_merge[0]
    for df in dfs_to_merge[1:]:
        merged_master_df = pd.merge(
            merged_master_df,
            df,
            on='Date',
            how='outer'
        )

    # Sort by date
    merged_master_df = merged_master_df.sort_values('Date').reset_index(drop=True)

    print(f"\nMerged dataframe shape: {merged_master_df.shape}")
    print(f"Date range: {merged_master_df['Date'].min()} to {merged_master_df['Date'].max()}")
    print(f"\nColumns: {list(merged_master_df.columns)}")
    print(f"\nFirst few rows:")
    print(merged_master_df.tail())
else:
    print("No valid dataframes to merge")
```

```python
import os

output_dir = '../data/merged_data'

output_file = os.path.join(output_dir, 'merged_master_df.csv')
merged_master_df.to_csv(output_file, index=False)
```
